## 抽样误差
由于抽样引起的样本的统计量与总体参数间的差异

个体差异与抽样。

### 标准误 SE
if $x \sim \mathcal{N}(\mu, \sigma^2)$, then $\bar{X} \sim \mathcal{N}(\mu, \sigma_{\bar{X}}^2)$.

further: $\sigma_{\bar{X}}^2 = \frac{\sigma^2}{n}$ is general, independent with types of distributions

反应样本均数的离散程度，也反应了抽样误差的大小。

### 中心极限定理

个体资料X服从偏态的分布，当样本量n较大时，样本均数$\bar{x}$近似服从正态分布。
可以来求样本均数的变异情况。

```julia
n = 10000
sample_std = zeros(n)
for i = 1:n
    pooled = rand(Normal(168.15,6),10000);
    sample_std[i] = std(pooled)
end
histogram(sample_std, title="std of samples")
```

# t分布
实际研究中，$\sigma$未知，用样本的标准差来替代。

$$ t = \frac{\bar{X} - \mu}{S / \sqrt{n}} $$

## 自由度
v = n-1

```julia
n = 7
sample_times = 10000

sample_u = zeros(sample_times)
sample_t = zeros(sample_times)
for i = 1:sample_times
    pooled = rand(Normal(168.15,6),n);
    sample_u[i] = (mean(pooled) - 168.15)/6*sqrt(n)
    sample_t[i] = (mean(pooled) - 168.15)/std(pooled)*sqrt(n)
end
histogram(sample_t,alpha=0.6,bins=100, xlim=(-5,5), label="u")
histogram!(sample_u,alpha=0.6,bins=100, xlim=(-5,5), label="t")
```

## 样本率的分布

样本率本身就是一个均数的概念。

e.g. 对于Binomial分布，$Var(p) = \frac{Var(x)}{n^2} $

$$\mu_P = \pi$$

$$\sigma_P = \sqrt{\frac{\pi (1-\pi)}{n}} $$

#### 均数差的可信区间
假定两组的方差是齐的，且数据相互独立，为正态。
合并方差：
$$ s_c^2 = \frac{(n_1-1)s_1^2 + (n_2-1s_2^2)}{n_1+n_2-2} $$

均数差的标准误：
$$ s_{\bar{X}_1 - \bar{X}_2} = \sqrt{s_c^2 (\frac{1}{n_1} - \frac{1}{n_2})} $$
