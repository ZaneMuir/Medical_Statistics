# 成组设计

完全随机分组和按不同人群抽样所得的样本均为独立样本资料。

## t检验
- 独立性
- 正态性 或样本量足够大
- 方差齐性

### 正态性检验
- Shapiro-Wilk检验
- 偏度与峰度联合检验

### 方差齐性检验
- F检验，通常这里设为 $\alpha = 0.1$

$$ F = \frac{S_{\mathrm{larger}}^2}{S_{\mathrm{smaller}}^2} $$

- 两组或多组可用Levene检验

## 双样本t检验 Student's t-test

### t值计算
$$ t = \frac{\bar{x_1} - \bar{x_2}}{s_{\bar{x_1} - \bar{x_2}}} = \frac{\bar{x_1} - \bar{x_2}}{\sqrt{s_c^2 (\frac{1}{n_1} + \frac{1}{n_2})}} $$

$$ s_c^2 = \frac{(n_1-1)s_1^2 + (n_2-1s_2^2)}{n_1+n_2-2} $$

$$ \nu = n_1 + n_2 - 2$$

## t'检验 Welch's t-test
方差不齐

### 计算t'值

$$ t' = \frac{\bar{x_1} - \bar{x_2}}{\sqrt{\frac{S_{\bar{x_1}}}{n_1} + \frac{S_{\bar{x_2}}}{n_2}}} $$

$$ \nu \approx \frac{(\frac{s_1^2}{N_1} + \frac{s_2^2}{N_2})^2}{\frac{s_1^4}{N_1^2\nu_1} + \frac{s_2^4}{N_2^2\nu_2}}$$


# 配对设计

- 同一样本接受不同处理的比较
- 同一对象治疗或者处理前后的比较
- 配对的两个受试对象分别给予两种处理

## 配对t检验
- 独立性: 对子间相互独立
- 正态分布性

## 计算t值

$$ t = \frac{\bar{d} - \mu_d}{S_{\bar{d}}} = \frac{\bar{d} - \mu_d}{\frac{S_d}{\sqrt{n}}}$$

$$ S_d = \sqrt{\frac{\sum d^2 - \frac{1}{n} (\sum d)^2}{n-1}} $$

$$ \nu = n-1 $$
